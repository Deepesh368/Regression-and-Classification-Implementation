# Logistic Regression with gradient descent:

```python
x = np.array(dataset[['Pregnancies', 'Glucose', 'BloodPressure','BMI', 'DiabetesPedigreeFunction', 'Age']].values)
y = np.array(dataset['Outcome']).reshape(-1,1)

for i in range(1,6):
    mean = np.mean(x[:,i])
    for j in range(len(x[:,i])):
        if(x[:,i][j] == 0):
            x[:,i][j] = mean

indices = np.array(range(len(x)))

num_training_indices = int(0.8 * len(x))
np.random.shuffle(indices)
train_indices = indices[0:num_training_indices]
test_indices = indices[num_training_indices: len(x)]

x_train, x_test = x[train_indices, :], x[test_indices, :]
y_train, y_test = y[train_indices, :], y[test_indices, :]

alpha = 0.0001
iterations = 10000

w = np.zeros(x_train.shape[1]).reshape(-1,1)
b = 0
m = len(x_train)
costs = []

for i in range(iterations):
    A = 1 / (1 + np.exp(-(np.dot(x_train, w)  + b)))

    num = A - y_train
    
    cost = np.sum(-y_train * np.log(A) - (1 - y_train) * np.log((1 - A))) * (1/m)
    costs.append(cost)
    
    dw = np.dot(x_train.T, num) / m
    db = np.sum(num) / m
    
    w = w - alpha * dw
    b = b - alpha * db
    
prediction = 1 / (1 + np.exp(-(np.dot(x_test, w)  + b)))
yhat = np.where(prediction > 0.5, 1, 0)

accuracy_score = 0
for i in range(len(y_test)):
    if(yhat[i] == y_test[i]):
        accuracy_score += 1
        

percent_correct = accuracy_score * 100 / len(y_test)   

plt.plot(costs)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

# Accuracy score for logistic regression with gradient descent

```python
print("Accuracy Score:" ,percent_correct)
```

# Logistic Regression with  Newtons's method:

```python
alpha = 1
iterations = 10000

w = np.zeros(x_train.shape[1]).reshape(-1,1)
b = 0
m = len(x_train)
costs = []

for i in range(iterations):
    A = 1 / (1 + np.exp(-(np.dot(x_train, w)  + b)))

    num = A - y_train
    
    cost = np.sum(-y_train * np.log(A) - (1 - y_train) * np.log((1 - A))) * (1/m)
    costs.append(cost)
    
    dw = np.dot(x_train.T, num) / m
    db = np.sum(num) / m
    
    H = np.dot((1 - A).T, A) * np.dot(x_train.T, x_train)  # Hessian for 'w'
    Hb = np.dot((1 - A).T, A)                              # Hessian for 'b'
    
    w = w - alpha * np.dot(np.linalg.inv(H), dw)
    b = b - alpha * (1 / Hb) * db
    
prediction = 1 / (1 + np.exp(-(np.dot(x_test, w)  + b)))
yhat = np.where(prediction > 0.5, 1, 0)

accuracy_score = 0
for i in range(len(y_test)):
    if(yhat[i] == y_test[i]):
        accuracy_score += 1
        

percent_correct = accuracy_score * 100 / len(y_test)   

plt.plot(costs)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

# Accuracy score for logistic regression with Newton's method:

```python
print("Accuracy Score:" ,percent_correct)
```
